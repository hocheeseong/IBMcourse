Question 2: Use Webscraping to Extract Tesla Revenue Data
Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the page containing revenue data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, "html5lib")

# Find the table containing revenue data (assuming it's the only table on the page)
table = soup.find("table")

# Extract the table rows
table_rows = table.find_all("tr")

# Create an empty list to store the data
data = []

# Iterate through the table rows and extract the data
for row in table_rows:
    cells = row.find_all("td")
    row_data = [cell.get_text(strip=True) for cell in cells]
    data.append(row_data)

# Create a DataFrame from the extracted data
df = pd.DataFrame(data)

# Display the last five rows using the tail function
print("Last five rows of Tesla Revenue DataFrame:")
print(df.tail())
```

Question 4: Use Webscraping to Extract GME Revenue Data
Display the last five rows of the gme_revenue dataframe using the tail function. Upload a screenshot of the results.



```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the page containing stock data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

# Find the table containing stock data (assuming it's the only table on the page)
table = soup.find("table")

# Extract the table rows
table_rows = table.find_all("tr")

# Create an empty list to store the data
data = []

# Iterate through the table rows and extract the data
for row in table_rows:
    cells = row.find_all("td")
    row_data = [cell.get_text(strip=True) for cell in cells]
    data.append(row_data)

# Create a DataFrame from the extracted data
df = pd.DataFrame(data)

# Display the last five rows using the tail function
print("Last five rows of GME DataFrame:")
print(df.tail())
```

Question 5: Plot Tesla Stock Graph

Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

Upload a screenshot of your results.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Load Tesla stock data
Tesla = yf.Ticker("TSLA")
Tesla_share_price_data = Tesla.history(period="max")
Tesla_share_price_data.reset_index(inplace=True)

# Define the make_graph function
def make_graph(data, x_col, y_col, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data[x_col], data[y_col], label=y_col, color='blue')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the make_graph function to plot Tesla stock data
make_graph(Tesla_share_price_data, 'Date', 'Close', 'Tesla Stock Price Over Time')
```
